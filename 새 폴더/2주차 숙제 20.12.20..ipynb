```python
def changeAge(n):
    if n < 10:
        return '유아'
    elif n < 20:
        return '10대'
    elif n < 30:
        return '20대'
    elif n < 40:
        return '30대'
    elif n < 50:
        return '40대'
    elif n < 60:
        return '50대'
    elif n < 70:
        return '60대'
    else:
        return '노인'
```

```python
titanic["Age_Range"] = titanic["Age"].apply(changeAge)
```

```python
titanic.groupby("Age_Range")["Survived"].sum()
```

```python
result = titanic.groupby('Age_Range')['Survived'].sum()
```

```python
result
```

```python
result = result.reindex(['유아','10대','20대','30대','40대','50대','60대','노인'])
result
```

```python
result_survived = titanic.groupby('Sex')['Survived'].sum()
result_survived
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.bar(result.index, result.values, color='m')
plt.title("나이대별 타이타닉 생존자")
plt.xlabel("나이")
plt.ylabel("생존자")

plt.subplot(122)
plt.bar(result_survived.index, result_survived.values, color='c')
plt.title("성별 타이타닉 생존자")
plt.xlabel("성별")
plt.ylabel("생존자")

plt.show()
```

```python
kr_daily = pd.read_csv('kr_daily.csv',header=0)
kr_daily
```

```python
def change_date(n):
    n = str(n)
    return n[:4] + "-" + n[4:6] + "-" + n[6:]
```

```python
kr_daily.index = kr_daily['date'].apply(change_date)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.xticks(rotation=80)
plt.title("확진자")
plt.plot(kr_daily["2020-05-18":]['confirmed'].index, kr_daily["2020-05-18":]['confirmed'].values, ":r")

plt.subplot(122)
plt.xticks(rotation=80)
plt.title("사망자")
plt.bar(kr_daily["2020-05-18":]['death'].index, kr_daily["2020-05-18":]['death'].values)
```
