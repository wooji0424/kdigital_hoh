```python
url = "https://search.musinsa.com/category/001?device=&d_cat_cd=001&brand=&rate=&page_kind=search&list_kind=small&sort=pop&sub_sort=&page=()&display_cnt=90&sale_goods=&ex_soldout=&color=&price1=&price2=&exclusive_yn=&size=&tags=&sale_campaign_yn=&timesale_yn=&q="

for i in range(1,11):
    response = requests.get(url.format(i))
    soup = BeautifulSoup(response.text, "html.parser")
    li_list = soup.select("#searchList li")
    for li in li_list:
        p_list = li.select("div.article_info  p")
        if p_list:
            print("브랜드:" ,p_list[0].text.strip())
            print("제품명:" ,re.sub(r"\[.*?\]","",p_list[1].text.strip()).strip())
            print("가격:" ,p_list[2].text.strip().split()[-1])
            print()
```
