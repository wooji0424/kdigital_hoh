```python
for page in range(1,6,1):
    url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=175318&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="+str(page)
    res = req.get(url)
    soup = bs(res.text,"lxml")
    
    reples = soup.select("div.score_reple > p")
    
    icon = soup.select("span.ico_viewer")
    
    for i in icon:
        i.extract()
        
    for index in reples:
        print(index.text.strip())
```

```python
from tqdm.notebook import tqdm
```

```python
# tqdm 진행사항을 알 표현할 수 있도록 해준다.
```

```python
replelist = []

for page in tqdm(range(1,51,1)):
    url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=175318&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="+str(page)
    res = req.get(url)
    soup = bs(res.text,"lxml")
    
    reples = soup.select("div.score_reple > p")
    
    icon = soup.select("span.ico_viewer")
    
    for i in icon:
        i.extract()
        
    for index in reples:
        print(index.text.strip())
        replelist.append(index.text.strip())
```

```python
replelist
```

```python
info = {"review": replelist}
```

```python
data = pd.DataFrame(info)
data
```

```python
data.set_index("review",inplace = True)
data
```

```python
data.to_csv("review.csv", encoding="euc-kr")
```
