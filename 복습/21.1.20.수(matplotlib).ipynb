## 실습1(타이타닉 생존자)

```python
#라이브러리
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#데이터 불러오기
titanic=pd.read_csv("titanic_train.csv")
titanic

#결측값 채우기
titanic["Age"]
titanic.fillna(titanic["Age"].mean(),inplace = True)
titanic["Age"].tail()
#결측값 확인하기
print(titanic["Age"].isnull().sum())

#새로운 데이터 만들기
def Agechange(n):
    if n < 10:
        return "유아"
    elif n < 20:
        return "10대"
    elif n < 30:
        return "20대"
    elif n < 40:
        return "30대"
    elif n < 50:
        return "40대"
    elif n < 60:
        return "50대"
    elif n < 70:
        return "60대"
    else:
        return "노인"

#데이터 적용하기
titanic["Age_Range"] = titanic["Age"].apply(Agechange)
result = titanic.groupby(titanic["Age_Range"])["Survived"].sum()
result = result.reindex(["유아","10대","20대","30대","40대","50대","60대","노인"])
result

result2 = titanic.groupby(titanic["Sex"])["Survived"].sum()
result2

#그래프 그리기
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.bar(result.index,result.values,color="m")
plt.title("나이별 타이타닉 생존자")
plt.xlabel("나이")
plt.ylabel("생존자")

plt.subplot(122)
plt.bar(result2.index, result2.values, color="blue")
plt.title("나이대별 타이타닉 생존자")
plt.xlabel("성별")
plt.ylabel("생존자")
```

## 실습2(코로나 데이터)

```python
# 데이터불러오기
kr_daily = pd.read_csv("kr_daily.csv")
kr_daily

# 새로운데이터 만들기
def chageDate(d):
    d = str(d)
    return d[0:4]+"-"+d[4:6]+"-"+d[6:8]

kr_daily["date1"] = kr_daily["date"].apply(chageDate)
kr_daily["date1"]= kr_daily["date"].astype("str")
kr_daily

# kr_daily.groupby(kr_daily["date1"])["confirmed"].sum()

kr_daily.index = kr_daily['date'].apply(chageDate)
kr_daily.columns

plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.xticks(rotation=80)
plt.title("확진자")
plt.plot(kr_daily["2020-05-18":]['confirmed'].index, kr_daily["2020-05-18":]['confirmed'].values, ":r")

plt.subplot(122)
plt.xticks(rotation=80)
plt.title("사망자")
plt.bar(kr_daily["2020-05-18":]['death'].index, kr_daily["2020-05-18":]['death'].values, color="m")
```
