```python
import re
```

```python
numbers = """
010-2334-3234
02-302-3033
010-1321-4043
02-01-32
03-3303-3033
016-444-3042
"""
```

```python
result = re.findall(r"[0-9]{3}-[0-9]{3,4}-[0-9]{4}",numbers)
result
```

```python
for result in numbers:
    print(result)
```

```python
results = re.finditer(r"[0-9]{3}-[0-9]{3,4}-[0-9]{4}",numbers)
results
```

```python
for result in results:
    print(result.group())
```

```python
email = re.search(r"[a-zA-z].@*.[a-zA-z]+\.[a-zA-Z]+",emails)
email #필요없는 부분
```

```python
for email in emails.split("\n"):
    matched = re.search(r"[a-zA-z].*@.[a-zA-z]+\.[a-zA-Z]+",email)
    if matched:
        print(matched.group())
```
