```python
urls = '''http://www.naver.com/
https://mail.naver.com/
ftp://ftp.naver.com/'''
```

```python
url_list = urls.split()
for url in url_list:
    protocol = re.match(r'.+\://', url).group()
    print(protocol)
    protocol = protocol[:-3]
    print(protocol)
```

```python
for url in url_list:
    protocol = re.match(r'.+(?=\://)', url).group()
    print(protocol)
```
