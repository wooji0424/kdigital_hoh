```python
def changeAge(n):
    if n < 10:
        return '유아'
    elif n < 20:
        return '10대'
    elif n < 30:
        return '20대'
    elif n < 40:
        return '30대'
    elif n < 50:
        return '40대'
    elif n < 60:
        return '50대'
    elif n < 70:
        return '60대'
    else:
        return '노인'
    
```

```python
titanic['Age_Range'] = titanic['Age'].apply(changeAge)
```

```python
titanic[['Age', 'Age_Range']].head()
```

```python
titanic.groupby('Age_Range')['Survived'].sum()
```
