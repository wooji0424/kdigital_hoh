## 3. 데이터 병합

```python
train = pd.merge(df, df1, on='custid', how='left')
train
```

## 4. 테이터 그룹핑
  - 구매자와 구매액
  - 환불한 인원
  - 구매자당 환불 횟수
  - 지점별 환불 횟수

 #### (1) 구매자와 구매액

```python
train_buy = train.groupby(["custid","amount"]).sum()
train_buy
```

#### (2) 환불한 인원 및 지점

```python
train_refund = train[train['amount']<0]
train_refund
```

#### (3) 고객당 환불 횟수

```python
train_refund1 = train.groupby(train[train['amount']<0]["custid"]).count()
train_refund1
```

```python
train_refund2 = train_refund.groupby(["custid","store"]).count()
train_refund2
```

## 5. 시각화

```python
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family='NanumGothic')

ratio = train_refund5['amount']/train_refund5['amount'].sum()
labels = ['본점', '무역점', '신촌점', '천호점']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

title = "지점별 환불 비율"
plt.title(title)
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

```python
plt.rc('font', family='NanumGothic')

ratio = train_refund5['amount']/train_refund5['amount'].sum()
labels = ['본점', '무역점', '신촌점', '천호점']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
# wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

title = "지점별 환불 비율"
plt.figure(figsize=(10,7))
plt.rcParams['font.size'] = 18
plt.title(title)
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```
