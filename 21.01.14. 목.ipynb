```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from IPython.core.display import Image
import numpy as np
import graphviz
import pydot
import pydotplus
import os

iris = load_iris()
iris.keys()

X = iris.data[:,:-2] #특징행열(데이터)과 대상 벡터(레이블) 만들기
y = iris.target


np.unique(iris.target, return_counts = True)

tree = DecisionTreeClassifier(max_depth = 2) #기울기와 절편을 가지고 있다.
tree.fit(X,y)

export_graphviz(tree, out_file="iris.dot",
                           feature_names=iris.feature_names[:-2],
                           class_names=iris.target_names, #분류
                           rounded=True, 
                           filled =True, #색상
                           impurity= True) #gini계수




with open("iris.dot") as f:
    dot_graph =f.read()
    
graphviz.Source(dot_graph)
```

```python
import pandas as pd
train_df = pd.read_csv("train.csv", nrows=1000)
train_df
```

```python
# 훈련용 데이터 만들기
```

```python
unused_columns, label_column = ["id", "hour", "device_id", "device_ip"],"click"
```

```python
train_df = train_df.drop(unused_columns, axis=1)
train_df
```

```python
aa = train_df.drop(unused_columns, axis=1)
aa
```

```python
X_dict_train = list(train_df.drop(label_column, axis=1).T.to_dict().values())
# 행과 열을 변환 T = transform
X_dict_train
```

```python
X_dict_train[0]
```

```python
# test 데이터 만들기
```

```python
y_train = train_df[label_column]
y_train
```

```python
test_df = pd.read_csv("train.csv", header=0, skiprows=(1, 1000),nrows=1000)
test_df
```

```python
test_df = test_df.drop(unused_columns, axis = 1)
test_df
```

```python
X_dict_test = list(test_df.drop(label_column, axis = 1).T.to_dict().values())
X_dict_test
```

```python
y_test = test_df[label_column]
y_test
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
vectorizer = DictVectorizer(sparse=True) # 희소해렬이 만들어짐
```

```python
X_train = vectorizer.fit_transform(X_dict_train)
X_train
```

```python
dir(vectorizer)
```

```python
X_train.shape
```

```python
X_train.toarray()[:5]
```

```python
X_test = vectorizer.fit_transform(X_dict_test)
X_test
```

```python
## X_train, y_train
## X_test, y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
parameters = {"max_depth": [3,10, None]}
```

```python
decision_tree = DecisionTreeClassifier(criterion="gini",min_samples_split = 30)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(decision_tree, parameters, n_jobs =-1, cv=3, scoring="roc_auc")
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
decision_tree_best = grid_search.best_estimator_
decision_tree_best
```

```python
#사이즈 조절 찾기
export_graphviz(decision_tree_best, out_file="click.dot",
                           feature_names=vectorizer.feature_names_,
                           class_names=["0","1"], #분류
                           rounded=True, 
                           filled =True, #색상
                           impurity= True) #gini계수



# import graphviz
with open("click.dot") as f:
    dot_graph =f.read()
# dot = graphviz.Source(dot_graph)
# dot.format = 'png'
# dot.render(filename='click', directory='./', cleanup=True)
graphviz.Source(dot_graph)
# graph.set_size('"10,10!"')
# graph.write_png("click")
```

```python
y_pred = decision_tree_best.predict(X_test)
print("y_pred = ", y_pred)
```

```python
np.unique(y_pred, return_counts=True)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred) # 정확도
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred) # 혼동행렬
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
y_pred_proba = decision_tree_best.predict_proba(X_test)[:,1]
y_pred_proba #클릭을 할 확률과 안할 확률 중 클릭할 확률만 추출
```

```python
fpr,tpr,_ = roc_curve(y_test, y_pred_proba) #roc커브 계산// ,_ 필수
auc = roc_auc_score(y_test,y_pred_proba) # roc커브의 auc계산
```

## ROC커브의 시각화
        - 모든 케이스에 대해 정확히 분류할 경우 (TPR=1,FPR=0) AUC면적은 1이 됨           (사각형 모형)

```python
import matplotlib.pyplot as plt
plt.plot(fpr,tpr,"r-",label = "DecisionTreeClassifier")
plt.plot([0,1],[0,1],"b--",label = "random guess")
plt.xlabel("false positive rate")
plt.ylabel("true positive rate")
plt.title("AUC={0:.2f}".format(auc))
plt.legend(loc = "lower right") 
```
