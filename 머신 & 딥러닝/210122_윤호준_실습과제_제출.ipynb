```python
# 1. 라이브러리

# 2. 데이터 수집
data2 = pd.read_csv("diabetes.csv")
x_train2 = np.array(data2[0:759].drop(columns='Outcome'))
y_train2 = np.array(data2[0:759]['Outcome'])
x_test2 = np.array(data2[759:].drop(columns="Outcome"))
y_test2 = np.array(data2[759:]['Outcome'])

# 3. 모델 구성
model = Sequential()
model.add(Dense(64, input_dim=8, activation='relu'))
model.add(Dense(1, input_dim=8, activation='sigmoid'))

model.compile(optimizer='adam' ,loss='binary_crossentropy',metrics=['binary_accuracy'])

# 4. 검증 세트 만들기
x_val = x_train2[:300]
partial_x_train2 = x_train2[300:]
y_val = y_train2[:300]
partial_y_train2 = y_train2[300:]

# 5. 모델 훈련
history = model.fit(partial_x_train2, partial_y_train2, epochs=100,
                   batch_size = 1, validation_data = (x_val, y_val),
                   verbose = 1)

# 6. 모델 평가
results = model.evaluate(x_test2, y_test2, verbose=2)
print(results)

# 7. 정확도 및 손실 그래프 도시
history_graph = history.history  #훈련 과정의 정보가 담겨 있음

  #손실 그래프
acc = history_graph['binary_accuracy']
val_acc = history_graph['val_binary_accuracy']
loss = history_graph['loss']
val_loss = history_graph['val_loss']
epochs = range(1, len(acc) + 1)
plt.plot(epochs, loss, 'bo', label="Training loss")
plt.plot(epochs, val_loss, 'b', label="Validation loss")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

  #정확도 그래프
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
